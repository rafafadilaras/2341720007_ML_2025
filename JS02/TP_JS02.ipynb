```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dpath = 'wbc.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.head()
```

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
std = StandardScaler()
df[df.columns.difference(['diagnosis'])] = std.fit_transform(df[df.columns.difference(['diagnosis'])])
```
